Preproccesing the Data 

```python
import pandas as pd

data = pd.read_csv('pokemon.csv', encoding='UTF-16', delimiter='\t')

# Updated list of columns to be removed
columns_to_remove = [
    'national_number', 'english_name', 'japanese_name', 'description',
    'is_sublegendary', 'is_legendary', 'is_mythical',
    'gigantamax', 'mega_evolution', 'mega_evolution_alt',
    'evochain_0', 'evochain_1', 'evochain_2', 'evochain_3', 'evochain_4', 'evochain_5', 'evochain_6',
    'abilities_0', 'abilities_1', 'abilities_2', 'abilities_hidden',
    'against_normal', 'against_fire', 'against_water',
    'against_electric', 'against_grass', 'against_ice', 'against_fighting',
    'against_poison', 'against_ground', 'against_flying', 'against_psychic',
    'against_bug', 'against_rock', 'against_ghost', 'against_dragon',
    'against_dark', 'against_steel', 'against_fairy'
]

data_cleaned = data.drop(columns=columns_to_remove)

data_cleaned.to_csv('cleaned_pokemon.csv', index=False, encoding='UTF-16')

data_cleaned.head()

```

Data Understanding/Visualization

```python
import pandas as pd

data = pd.read_csv('cleaned_pokemon.csv', encoding='UTF-16')

```

```python
# Get summary statistics
summary = data.describe()
print(summary)
```

Visualizing Distributions

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

sns.histplot(data['hp'])
plt.show()
```

Correlation Matrix

```python
numeric_data = data.select_dtypes(include=['number'])

corr = numeric_data.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, fmt=".2f")
plt.show()
```

Class Distributions for primary type

```python
sns.countplot(y=data['primary_type'], order=data['primary_type'].value_counts().index)
plt.show()
```
